## Descripción de la tablas

Variable and File Structure

The file structure is the same for all downloaded files. Events are atomic, in that they are recorded by  day and location, actor and event type. For example, if a battle between a government and rebel group  occurs over three days in 
one town, this is recorded as three separate events in ACLED. If during that  time, there are reports of violence against civilians in the same town, this is recorded as a separate event, as the actors and event type are different. Data can be aggregated  by month, regions, or event types using pivot tables.

Se probó con la base de datos en formato csv y dado que algunas las celdas contenían texto en el que utilizaban comas, representaba muchos problemas. Se optó por tomar la base en formato xlsx para mantener la correcta separación de las columnas.

```python
import csv
import xlrd

base = 'data/data_z.xlsx'

def convertDateToString(value, datemode):
    year, month, day, hour, minute, second = xlrd.xldate_as_tuple(value,0)
    return '{}/{:02}/{:02}'.format(year,month,day)
    
workbook = xlrd.open_workbook(base)
sheet = workbook.sheet_by_index(0)

with open('data/data_w.psv', 'w') as outfile:
    writer = csv.writer(outfile, delimiter='Þ')
    for i in range(sheet.nrows):
        writer.writerow([cell.value if cell.ctype != 3 else convertDateToString(cell.value,workbook.datemode) for cell in sheet.row(i)])
```

## Carga de información a Spark

```python
import pyspark
sc = pyspark.SparkContext('local[*]')
sqlContext = pyspark.SQLContext(sc)
from pyspark.sql import Row
```

```python
Eventos = Row('GWNO','EVENT_ID_CNTY','EVENT_ID_NO_CNTY','EVENT_DATE','YEAR','TIME_PRECISION','EVENT_TYPE','ACTOR1','ALLY_ACTOR_1','INTER1','ACTOR2','ALLY_ACTOR_2','INTER2','INTERACTION','COUNTRY','ADMIN1','ADMIN2','ADMIN3','LOCATION','LATITUDE','LONGITUDE','GEO_PRECISION','SOURCE','NOTES','FATALITIES')
```

```python
def getEventos(linea):
    cells = linea.split('Þ')
    return Eventos(*cells)
```

```python
eventos_rdd = sc.textFile("data/data_y.psv")
```

```python
eventos_rdd.take(5)
```

```python
header = eventos_rdd.first()
eventos = eventos_rdd.filter(lambda line: line != header)\
                     .map(getEventos)\
                     .toDF()
        
eventos.take(5)
```

## Funciones dentro de SQL

```python
#Tenemos 117,823 ataques registrados
eventos.count()
```

```python
#Cómo extraemos el mes?
print('1997/01/11'[5:7])
#Cómo extraemos el día?
print('1997/01/11'[8:10])
```

```python
from pyspark.sql.types import IntegerType
from pyspark.sql.types import StringType

def fatalities(FATALITIES):
    return int(FATALITIES)

def month_of_event(EVENT_DATE):
    return int(EVENT_DATE[5:7])

def day_of_event(EVENT_DATE):
    return int(EVENT_DATE[8:10])
```

```python
sqlContext.registerFunction('fatalities', fatalities, IntegerType())
sqlContext.registerFunction('month_of_event', month_of_event, IntegerType())
sqlContext.registerFunction('day_of_event', day_of_event, IntegerType())
```

```python
eventos.registerTempTable('eventos')
```

```python
eventos_clean = sqlContext.sql("select GWNO as GWNO,EVENT_ID_CNTY as EVENT_ID_CNTY ,EVENT_ID_NO_CNTY as EVENT_ID_NO_CNTY,YEAR as YEAR,TIME_PRECISION as TIME_PRECISION,EVENT_TYPE as EVENT_TYPE,ACTOR1 as ACTOR1,ALLY_ACTOR_1 as ALLY_ACTOR_1,INTER1 as INTER1,ACTOR2 as ACTOR2,ALLY_ACTOR_2 as ALLY_ACTOR_2,INTER2 as INTER2,INTERACTION as INTERACTION,COUNTRY as COUNTRY,ADMIN1 as ADMIN1,ADMIN2 as ADMIN2,ADMIN3 as ADMIN3,LOCATION as LOCATION,LATITUDE as LATITUDE,LONGITUDE as LONGITUDE, 'GEO_PRECISION' as GEO_PRECISION,'SOURCE' as SOURCE,'NOTES' as NOTES, fatalities(FATALITIES) as FATALITIES, month_of_event(EVENT_DATE) as MONTH_OF_EVENT, day_of_event(EVENT_DATE) AS DAY_OF_EVENT from eventos")
```

```python
eventos_clean.registerTempTable('eventos_clean')
```

## Somalia, el país con mayor número de conflictos armados


INTER1 = 2 representa a fuerzas rebeldes e INTER1 = 1 representa a los gobiernos

```python
import matplotlib.pyplot as plt

labels = (np.array(df4['COUNTRY']))[0:5]
labels_other = (np.array(df4['COUNTRY']))[5:]
total = sum(df4['Events'])
fracs = ([float(i) for i in np.array(df4['Events'])][0:5]) / total
fracs_other = ([float(i) for i in np.array(df4['Events'])][5:]) / total

labels = np.concatenate((labels[0:5], ['Other']), axis=0)
fracs = np.concatenate((fracs[0:5], [sum(fracs_other)]), axis=0)
explode = [0]*len(fracs)
colors = ['yellowgreen', 'mediumpurple', 'lightskyblue', 'lightcoral', 'pink', 'yellow']
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
plt.pie(fracs, explode=explode, labels=labels,autopct='%1.1f%%', colors=colors, startangle=90)
plt.axis('equal')
plt.title('Principales países con mayor porcentaje de conflictos durante 2015',fontsize=14, y=1.08)
plt.show()
```
